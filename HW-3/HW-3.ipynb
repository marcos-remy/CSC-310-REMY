```python
import pandas as pd
from treeviz import tree_print
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("tennis_numeric.csv")

print(data.columns)

# Assuming X contains the features and y contains the target variable
X = data.drop(columns=['play'])
y = data['play']

# Create and train the decision tree
dt = DecisionTreeClassifier()
dt.fit(X, y)

# Visualize the decision tree
tree_print(dt, X)

# Model Accuracy
y_pred = dt.predict(X)
accuracy = accuracy_score(y, y_pred)
print(f"Accuracy on training data: {accuracy}")
print("Model has perfect accuracy")

dt1 = DecisionTreeClassifier(max_depth=2)
dt1.fit(X, y)

# Visualize the decision tree
tree_print(dt1, X)


print("In both versions of the tree, the outlook being overcast is a strong predictor of playing tennis. Humidity also plays a role in the decision-making process, particularly when the outlook is not overcast. Temperature is considered in the more complex tree but not in the pruned tree.")
```

```python
# Load the dataset
data = pd.read_csv("caesarian.csv")

print(data.columns)

# Assuming X contains the features and y contains the target variable
X = data.drop(columns=['Caesarian'])
y = data['Caesarian']

# Create and train the decision tree
dt = DecisionTreeClassifier()
dt.fit(X, y)

# Visualize the decision tree
tree_print(dt, X)

# Model Accuracy
y_pred = dt.predict(X)
accuracy = accuracy_score(y, y_pred)
print(f"Accuracy on training data: {accuracy}")

# Model has perfect accuracy
print("Model has really good accuracy")

# Prune the tree with max_depth=2
dt_pruned = DecisionTreeClassifier(max_depth=3)
dt_pruned.fit(X, y)

# Visualize the pruned decision tree
tree_print(dt_pruned, X)

# Provide interpretations of the patterns
print("The first decision is based on whether there is a "Heart Problem" or not. If there is no heart problem (Heart Problem =< 0.5), the model looks at the persons Age. If the age is less than or equal to 21.5, it further refines the decision based on age. If the age is less than or equal to 18.5, the prediction is True (This may indicate a certain outcome). If the age is greater than 18.5, the prediction is False (This may indicate a different outcome). If the age is greater than 21.5, the model looks at the Delivery number. If the delivery number is less than or equal to 2.5, the prediction is True. If the delivery number is greater than 2.5, the prediction is False. If there is a heart problem (Heart Problem > 0.5), the model looks at the Blood of Pressure. If the blood pressure is less than or equal to 0.5, the prediction is True. If the blood pressure is greater than 0.5, the model looks at the Delivery time. If the delivery time is less than or equal to 0.5, the prediction is False. If the delivery time is greater than 0.5, the prediction is True")
```
