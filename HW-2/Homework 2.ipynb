```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your dataset
df = pd.read_csv('BankWages.csv')
```

```python
# 1. How many columns are there?
num_columns = len(df.columns)
print(f"Number of columns: {num_columns}")
```

```python
# 2. How many rows are there?
num_rows = len(df)
print(f"Number of rows: {num_rows}")
```

```python
#3. Are the columns numerical, categorical, or both?

# Get a summary of column types
column_types = df.dtypes

# Determine column types
numerical_columns = column_types[column_types != 'object'].index.tolist()
categorical_columns = column_types[column_types == 'object'].index.tolist()

#Display column types
print(numerical_columns)
print(categorical_columns)

#Answer:
if numerical_columns and categorical_columns:
    print("Both")
elif numerical_columns:
    print("Numerical")
elif categorical_coolumns:
    print("Categorical")
```

```python
#4. Are there columns that provide no information? Just constant or unique values for all observations.
# Identify columns with constant values
constant_columns = df.columns[df.nunique() == 1]

# Identify columns with unique values for all observations
unique_columns = df.columns[df.nunique() == len(df)]

# Print the results
if len(constant_columns) > 0 or len(unique_columns) > 0:
    print("Yes, there are columns that provide no information.")
else:
    print("No, there aren't any columns that provide no information.")
```

```python
#5. For each categorical columns compute the mode.

# Identify categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Compute the mode for each categorical column
modes = df[categorical_columns].mode()

# Print the modes
print("Mode for each categorical column:")
print(modes)
```

```python
#6. For each numerical column compute the mean (average), max, and min values.

# Identify numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns

# Compute statistics for each numerical column
statistics = df[numerical_columns].agg(['mean', 'max', 'min'])

# Print the statistics
print("Statistics for each numerical column:")
print(statistics)
```

```python
#7. Plot histograms for numerical columns -- comment on the distribution (normal, multi-modal etc)

# Identify numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns

# Plot histograms for numerical columns
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#8. Plot bar plots for categorical columns. -- comment on whether the data set is balanced in terms of the representation of the various labels in the categorical columns.

# Plot bar plots for categorical columns
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    df[column].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f'Bar Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
    
    # Comment on dataset balance
    num_categories = len(df[column].value_counts())
    total_entries = len(df[column])
    balance_ratio = num_categories / total_entries
    
    if balance_ratio < 0.2 or balance_ratio > 5:
        print(f"The dataset might be imbalanced for the column '{column}'.")
    else:
        print(f"The dataset seems balanced for the column '{column}'.")
```
